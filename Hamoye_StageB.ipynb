```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('C:\\Users\\omoni\\OneDrive\\Desktop\\Practice\\Hamoye AI\\energydata_complete.csv')
df.head(2)
```

```python
df['year'] = [item[0:4] for item in df['date'].values]
df['month'] = [item[5:7] for item in df['date'].values]
df['day'] = [item[8:10] for item in df['date'].values]
df['time'] = [item[11:16] for item in df['date'].values]
df.drop(columns=['date'], inplace=True)
df.head()
```

```python
Q1_df = df[['T2', 'T6']]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(Q1_df), columns=Q1_df.columns)
features_df = normalised_df.drop(columns=['T6'])
heating_target = normalised_df['T6']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, heating_target, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
predicted_values = linear_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score,3)
```

```python
df.drop(columns=['date'], inplace=True)
df.head()
```

```python
df.drop(columns=['lights'], inplace=True)
```

```python
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
features_df = normalised_df.drop(columns=['Appliances'])
heating_target = normalised_df['Appliances']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, heating_target, test_size=0.3, random_state=42)
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
predicted_values = linear_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae,3)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse,3)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score,3)
```

```python
print(' residual sum of squares is : '+ str(np.sum(np.square( predicted_values-y_test ))))
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.5)
ridge_reg.fit(x_train,y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)

#comparing the effects of regularisation 
def get_weights_df(model, feat, col_name):
    #this feature returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['features', col_name]
    weights_df[col_name].round(3)
    return weights_df

linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')

final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='features')

final_weights
```
